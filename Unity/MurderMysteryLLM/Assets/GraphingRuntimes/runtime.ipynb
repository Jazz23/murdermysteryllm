```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

**Load Data**

```python
p2_ollama_00 = pd.read_csv("csv/player_count_2/ollama_runtime_trial_01.csv")
p2_ollama_01 = pd.read_csv("csv/player_count_2/ollama_runtime_trial_02.csv")
p2_ollama_02 = pd.read_csv("csv/player_count_2/ollama_runtime_trial_03.csv")

player_2_ollama = pd.concat([p2_ollama_00, p2_ollama_01, p2_ollama_02], ignore_index=True)

p3_ollama_00 = pd.read_csv("csv/player_count_3/ollama_runtime_trial_01.csv")
p3_ollama_01 = pd.read_csv("csv/player_count_3/ollama_runtime_trial_02.csv")
p3_ollama_02 = pd.read_csv("csv/player_count_3/ollama_runtime_trial_03.csv")
player_3_ollama = pd.concat([p3_ollama_00, p3_ollama_01, p3_ollama_02], ignore_index=True)

p4_ollama_00 = pd.read_csv("csv/player_count_4/ollama_runtime_trial_01.csv")
p4_ollama_01 = pd.read_csv("csv/player_count_4/ollama_runtime_trial_02.csv")
p4_ollama_02 = pd.read_csv("csv/player_count_4/ollama_runtime_trial_03.csv")
player_4_ollama = pd.concat([p4_ollama_00, p4_ollama_01, p4_ollama_02], ignore_index=True)

p5_ollama_00 = pd.read_csv("csv/player_count_5/ollama_runtime_trial_01.csv")
p5_ollama_01 = pd.read_csv("csv/player_count_5/ollama_runtime_trial_02.csv")
p5_ollama_02 = pd.read_csv("csv/player_count_5/ollama_runtime_trial_03.csv")
player_5_ollama = pd.concat([p5_ollama_00, p5_ollama_01, p5_ollama_02], ignore_index=True)
```

```python
p2_task_00 = pd.read_csv("csv/player_count_2/task_response_time_trail_01.csv")
p2_task_01 = pd.read_csv("csv/player_count_2/task_response_time_trial_02.csv")
p2_task_02 = pd.read_csv("csv/player_count_2/task_response_time_trial_03.csv")
player_2_task = pd.concat([p2_task_00, p2_task_01, p2_task_02], ignore_index=True)

p3_task_00 = pd.read_csv("csv/player_count_3/task_response_time_trial_01.csv")
p3_task_01 = pd.read_csv("csv/player_count_3/task_response_time_trial_02.csv")	
p3_task_02 = pd.read_csv("csv/player_count_3/task_response_time_trial_03.csv")
player_3_task = pd.concat([p3_task_00, p3_task_01, p3_task_02], ignore_index=True)

p4_task_00 = pd.read_csv("csv/player_count_4/task_response_time_trial_01.csv")
p4_task_01 = pd.read_csv("csv/player_count_4/task_response_time_trial_02.csv")
p4_task_02 = pd.read_csv("csv/player_count_4/task_response_time_trial_03.csv")
player_4_task = pd.concat([p4_task_00, p4_task_01, p4_task_02], ignore_index=True)

p5_task_00 = pd.read_csv("csv/player_count_5/task_response_time_trial_01.csv")
p5_task_01 = pd.read_csv("csv/player_count_5/task_response_time_trial_02.csv")
p5_task_02 = pd.read_csv("csv/player_count_5/task_response_time_trial_03.csv")
player_5_task = pd.concat([p5_task_00, p5_task_01, p5_task_02], ignore_index=True)
```

**Ollama Runtime**

```python
# Add a "Source" label to distinguish datasets
player_2_ollama['Source'] = '2 Agents'
player_3_ollama['Source'] = '3 Agents'
player_4_ollama['Source'] = '4 Agents'
player_5_ollama['Source'] = '5 Agents'

# Combine the datasets
combined_df = pd.concat([player_2_ollama, player_3_ollama, player_4_ollama, player_5_ollama], ignore_index=True)

reasoning_df = combined_df[combined_df['Task'] == 'Reasoning']
dialogue_df = combined_df[combined_df['Task'] == 'Dialogue']

# # Plot Reasoning Scatterplot
# plt.figure(figsize=(10, 6))
# sns.scatterplot(
#     data=reasoning_df,
#     x="CharacterCount",
#     y="ResponseTime(ms)",
#     hue="Source",
#     palette="viridis",
#     s=100,
#     edgecolor="black"
# )
# plt.title("Reasoning Task: Response Time vs. Character Count")
# plt.xlabel("Character Count")
# plt.ylabel("Response Time (ms)")
# plt.grid(True)
# plt.legend(title="Agent Count")
# plt.tight_layout()
# plt.show()

# # Plot Dialogue Scatterplot
# plt.figure(figsize=(10, 6))
# sns.scatterplot(
#     data=dialogue_df,
#     x="CharacterCount",
#     y="ResponseTime(ms)",
#     hue="Source",
#     palette="plasma",
#     s=100,
#     edgecolor="black"
# )
# plt.title("Dialogue Task: Response Time vs. Character Count")
# plt.xlabel("Character Count")
# plt.ylabel("Response Time (ms)")
# plt.grid(True)
# plt.legend(title="Agent Count")
# plt.tight_layout()
# plt.show()


# Plot 3: Average Response Time per Task by Agent Count
plt.figure(figsize=(10, 6))
sns.barplot(
    data=combined_df,
    x="Task",
    y="ResponseTime(ms)",
    hue="Source",
    estimator='mean',
    errorbar='sd',
    palette="coolwarm"
)
plt.title("Average Response Time by Task and Agent Count")
plt.xlabel("Task Type")
plt.ylabel("Avg Response Time (ms)")
plt.grid(axis='y')
plt.tight_layout()
plt.show()

# Plot 4: Distribution Plot for Reasoning
plt.figure(figsize=(10, 6))
sns.histplot(
    data=reasoning_df,
    x="ResponseTime(ms)",
    hue="Source",
    kde=True,
    bins=15,
    palette="viridis"
)
plt.title("Distribution of Reasoning Response Times")
plt.xlabel("Response Time (ms)")
plt.tight_layout()
plt.show()

# Plot 5: Distribution Plot for Dialogue
plt.figure(figsize=(10, 6))
sns.histplot(
    data=dialogue_df,
    x="ResponseTime(ms)",
    hue="Source",
    kde=True,
    bins=15,
    palette="plasma"
)
plt.title("Distribution of Dialogue Response Times")
plt.xlabel("Response Time (ms)")
plt.tight_layout()
plt.show()
```

```python
# Label each dataset
player_2_task['Source'] = '2 Agents'
player_3_task['Source'] = '3 Agents'
player_4_task['Source'] = '4 Agents'
player_5_task['Source'] = '5 Agents'

# Combine all into one DataFrame
combined_task_df = pd.concat([player_2_task, player_3_task, player_4_task, player_5_task], ignore_index=True)


# Clean column names
combined_task_df.columns = combined_task_df.columns.str.strip()
combined_task_df.columns = combined_task_df.columns.str.replace(" ", "")

# Print to verify
print("Final Columns:", combined_task_df.columns)

# Proceed with plot
plt.figure(figsize=(12, 6))
sns.barplot(
    data=combined_task_df,
    x="TaskType",  # confirm this matches your column name
    y="ResponseTime(ms)",
    hue="Source",
    estimator=np.mean,
    errorbar='sd',
    palette="coolwarm"
)
plt.title("Average Response Time by Task Type and Agent Count")
plt.ylabel("Avg Response Time (ms)")
plt.xlabel("Task Type")
plt.grid(axis='y')
plt.tight_layout()
plt.show()


#  Plot 2: TaskTime vs TotalTime (Faceted by Agent Count)
# g1 = sns.FacetGrid(
#     combined_task_df,
#     col="Source",
#     hue="TaskType",
#     col_wrap=2,
#     height=4,
#     aspect=1.3,
#     palette="Set2"
# )
# g1.map(sns.scatterplot, "TaskTime(ms)", "TotalTime(ms)", s=60, edgecolor="black")
# g1.add_legend()
# g1.fig.subplots_adjust(top=0.9)
# g1.fig.suptitle("TaskTime vs TotalTime by Agent Trial")
# plt.show()

#  Plot 3: Response Time Distribution (Faceted by Agent Count)
g2 = sns.FacetGrid(
    combined_task_df,
    col="Source",
    hue="TaskType",
    col_wrap=2,
    height=4,
    aspect=1.3,
    palette="mako"
)
g2.map(sns.histplot, "ResponseTime(ms)", bins=15, kde=True)
g2.add_legend()
g2.fig.subplots_adjust(top=0.9)
g2.fig.suptitle("LLM Response Time Distribution by Agent Trial")
plt.show()
```
